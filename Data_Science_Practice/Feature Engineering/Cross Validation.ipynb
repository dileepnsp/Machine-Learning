```python
import pandas as pd
import numpy as np
df=pd.read_csv('Datasets/Purchased_Dataset.csv')
df.columns
```

```python
x=df[['Age','EstimatedSalary']]
y=df['Purchased']
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=3)
knnclassifier=KNeighborsClassifier(n_neighbors=5)
knnclassifier.fit(x_train,y_train)
y_pred=knnclassifier.predict(x_test)
print(metrics.accuracy_score(y_test,y_pred))
x_train.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=2)
knnclassifier=KNeighborsClassifier(n_neighbors=5)
knnclassifier.fit(x_train,y_train)
y_pred=knnclassifier.predict(x_test)
print(metrics.accuracy_score(y_test,y_pred))
x_train.head()
```

```python
from sklearn.model_selection import cross_val_score
knnclassifier=KNeighborsClassifier(n_neighbors=5)
print(cross_val_score(knnclassifier,x,y,cv=10,scoring='accuracy'))
```

```python
print(cross_val_score(knnclassifier,x,y,cv=10,scoring='accuracy').mean())
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
print(cross_val_score(logreg,x,y,cv=10,scoring='accuracy'))
```
