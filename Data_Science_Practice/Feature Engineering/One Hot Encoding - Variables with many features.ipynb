```python
import pandas as pd
import numpy as np

data=pd.read_csv('../Datasets/benz.csv',usecols=['X1','X2','X3','X4','X5','X6'])
data.head()
```

```python
for col in data.columns:
    print(col,":",len(data[col].value_counts()))
```

```python
pd.get_dummies(data,drop_first=True)
```

```python
#Find top 10 most frequent categeries for X2 column
data['X2'].value_counts().sort_values(ascending=False).head(10)
```

```python
x2_top_10_categeries=data['X2'].value_counts().sort_values(ascending=False).head(10).index
x2_top_10_categeries
```

```python
for x in x2_top_10_categeries:
    data[x]=np.where(data.X2==x,1,0)
data[x2_top_10_categeries].head(40)
```
