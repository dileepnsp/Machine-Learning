```python
import pandas as pd
import statsmodels.api as sm
```

```python
df_adv=pd.read_csv('Datasets/Advertising.csv')
df_adv=df_adv.drop(['Unnamed: 0'],axis=1)
print(df_adv.columns)
```

```python
#x=df_adv.iloc[:,:-1]
x=df_adv[['TV','radio','newspaper']]
y=df_adv['sales']
y
```

```python
## Fit a OLS model with intercept on Tv and Radio
x=sm.add_constant(x)
model=sm.OLS(y,x).fit()

model.summary()
```

```python
x[['TV','radio','newspaper']].corr()
#There is no correlation among the features..So there is no Multicollineartiy here.
```
