```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
credits=pd.read_csv("tmdb_5000_credits.csv")
credits.head()
```

```python
credits.shape
```

```python
movies=pd.read_csv("tmdb_5000_movies.csv")
movies.head()
```

```python
movies.shape
```

```python
credits_column_renamed=credits.rename(columns={"movie_id":"id"})
```

```python
movies_df_merge=movies.merge(credits_column_renamed,on="id")
movies_df_merge.head()
```

```python
movies_df_merge.columns
```

```python
movies_df_merge.drop(columns=['budget', 'genres', 'homepage', 
        'keywords', 'original_language',
        'overview',  'production_companies',
       'production_countries', 'release_date', 'revenue', 'runtime',
       'spoken_languages', 'status', 'tagline', 'title_x','title_y', 'cast', 'crew'],axis=1,inplace=True)
```

```python
movies_df_merge.columns
```

# Using Weighted Average for each movies Average Rating
W=Rv+Cm/v+m

W=Weighted Rating

R=Average for the movie as a number from 0 to 10(mean) =Rating

v=no.of votes for the movie = votes

m = minimum votes required to be listed in the TOp 250 

C= the mean vote across the whole report 


```python
v=movies_df_merge['vote_count']
R=movies_df_merge['vote_average']
C=movies_df_merge['vote_average'].mean()
m=movies_df_merge['vote_count'].quantile(0.70)
```

```python
movies_df_merge['weighted_average']=((R*v)+(C*m))/(v+m)
```

```python
movies_df_merge.head()
```

```python
weighted_average=movies_df_merge.sort_values("weighted_average",ascending=False)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
sns.barplot(x=weighted_average['weighted_average'].head(10),y=weighted_average['original_title'].head(10))
plt.title("Top movies by Weighted Average")
plt.show()
```

```python
movies_popularity=movies_df_merge.sort_values("popularity",ascending=False)
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
sns.barplot(x=movies_popularity['popularity'].head(10),y=movies_popularity['original_title'].head(10))
plt.title("Top movies by Popularity")
plt.show()
```

# Recommendation Based on popularity 50% and weighted average 50%

```python
from sklearn.preprocessing import MinMaxScaler
scaling=MinMaxScaler()
movie_scaled_df=scaling.fit_transform(weighted_average[['weighted_average','popularity']])
movie_normalized_df=pd.DataFrame(movie_scaled_df,columns=['weighted_average','popularity'])
movie_normalized_df.head()
```

```python
weighted_average[['normalised_weighted_avg','normalised_popularity']]=movie_normalized_df
weighted_average.head()
```

```python
weighted_average['score']=weighted_average['normalised_weighted_avg']*0.5+weighted_average['normalised_popularity']*0.5
```

```python
weighted_averaged_sorted=weighted_average.sort_values(['score'],ascending=False)
weighted_averaged_sorted.head()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x=weighted_averaged_sorted['score'].head(10),y=weighted_averaged_sorted['original_title'].head(10))
plt.show()
```
